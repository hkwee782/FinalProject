## Intro: Submetric #1 - Number of Dogs
We estimated the number of dogs in the neighborhood based off the number of dog licenses in the neighborhood.

Clearly, dogs would do better in a neighborhood where there are plenty of other dogs to socialize with and where neighbors are used to having dogs around, but not too many or your dog might get into fights.

```python
import pandas as pd
```

```python
# import the dog license data
data = pd.read_csv("dogs_valid.csv")

# drop duplicates
data.drop_duplicates()

# display first 5 data points
data.head(5)
```

```python
# sort data by zip code of dog licenses, and presumably dogs
data.groupby("OwnerZip").count()
```

```python
# some areas don't seem to have a lot of dogs and some have too many, so we will filter those out
def min(x):
    return len(x) > 250

def max(x):
    return len(x) < 450

pop_zip = data.groupby("OwnerZip").filter(min)
avg_zip = pop_zip.groupby("OwnerZip").filter(max)

# display count by zip code again
zipCodes = avg_zip.groupby("OwnerZip").count()
zipCodes
```

```python
# sort high to low
sorted = zipCodes.sort_values("_id", ascending = False)
sorted
```

```python
# take the top zip codes
top = sorted.head(10)
# top
```

```python
# selecting count and zip code
top = top[top.columns[1]]
```

```python
# plot the chart
top.plot.bar(figsize=(20, 20))
```

## Conclusion - Submetric #1
The neighborhoods with an average population of dogs are Oakmont, Aspinwall and Sharpsburg, Dormont, McKeesport, Penn Hills, Elizabeth, White Oak, Swissvale, West End, and Etna.

## Intro: Submetric #2 - Size and Number of Parks

Pittsburgh is home to over 160 parks spanning thousands of acres, all that are great for our pets wellbeing. In this section we look specifically at the number of parks along with the total acreage of each neighborhood to determine which neighborhood is best for a dog.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Read in both shape files
neighborhoods = gpd.read_file("Neighborhoods/Neighborhoods_.shp")
parks = gpd.read_file("ParkShape/PGHWebParks.shp")

# Create plot with size 10
fig, ax = plt.subplots(figsize=(10, 10))

# Fix coordinate system
parks = parks.to_crs(neighborhoods.crs)

# Plots neighborhoods plot with colors first
neighborhoods.plot(ax=ax, color='lightgrey', edgecolor='black')

# Overlays parks plot over top
parks.plot(ax=ax, color='green', edgecolor='black', alpha=0.6)
plt.title('Neighborhoods and Parks')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#print(neighborhoods.crs)
#print(parks.crs)
```

```python
# Read in park csv and clean up unwanted columns
df = pd.read_csv("park.csv")
df = df.drop(['_id', 'objectid_1', 'objectid', 'final_cat', 'sector', 'shape_leng', 'globalid', 'created_user', 'created_date', 'last_edited_user', 'last_edited_date', 'maintenanceresponsibility', 'dpw_ac', 'globalid_1', 'shape_area', 'shape_length', 'alterntnam', 'divname'], axis=1)

# Count number of parks for each neighborhood
park_counts = df['Neighborhood'].value_counts()

# Sorts neighborhoods by most parks to least number of parks
top_5 = park_counts.nlargest(10)
top_5.plot(kind='bar', rot=0, figsize=(20, 20))
```

```python
# Counts the acreage by 
acreage_sums = df.groupby('Neighborhood')['acreage'].sum()

# Sorts by highest acreage values and gets top 5 values
ranked_acreage = acreage_sums.sort_values(ascending=False)
top_acreage = ranked_acreage.nlargest(10)

# Plot
top_acreage.plot(kind='bar', rot=0, figsize=(20,20))
```

## Conclusion: Submetric #2 

Based on the data above we can see that Squirrel hill is the best neighborhood statistically for a dog to live in based on a mixture of having one of the highest number of parks as well as by sheer area being one of the most predominant. Personally I believe Squirrel Hill is a great neighborhood alongside my personal choice which is one of the runner ups in the number of parks category, South Side.
